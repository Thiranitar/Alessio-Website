```python
from numpy.core.fromnumeric import size
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

#comandi da utente
theta = float(input("Inserisci il valore dell'angolo: "))
v0 = float(input("Inserisci il valore della velocità iniziale: "))
mass = float(input("Inserisci il valore della massa: "))
h0 = float(input("Inserisci il valore dell'altezza iniziale dal suolo: "))
a = float(input("Inserisci il valore dell'accellerazione di gravità: "))

v0x = v0 * np.cos(theta)
v0y = v0 * np.sin(theta)
print("Velocità iniziale orizzontale: ", v0x)
print("Velocità iniziale verticale: ",v0y)

tf = (v0y + np.sqrt(v0y**2 + 2*a*h0))/a
xf = v0x * tf
print("Gittata: ",xf)
print("Tempo di volo: ",tf)

t = np.arange(0,tf,0.01)
dimensionet = size(t)
print(dimensionet)
print(t)

xt = v0x * t
yt = v0y * t - 0.5 *a*t**2

vxt = np.zeros(282) + v0x
print(vxt) 
vyt= v0y - a*t
plt.plot(t,xt)
plt.show()
plt.plot(t,yt)
plt.show()
plt.plot(t,vxt)
plt.show()
plt.plot(t,vyt)
plt.show()

plt.plot(xt,yt)
plt.show()

ekin = 0.5 * mass * (vxt**2 +vyt**2)
plt.plot(t,ekin)
epot = mass*a*yt
etot = ekin + epot
plt.plot(t,epot)
plt.plot(t,etot)
plt.legend(["ekin","epot","etot"] , loc = "right")
plt.show()
```
